```python
import pandas as pd
df = pd.read_csv('prompts.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from PIL import Image

# Sample data for a row
data = df.loc[30] ## Change index to visualise

# Create a 1x3 grid for images and information
fig = plt.figure(figsize=(11, 3))
gs = gridspec.GridSpec(1, 4, width_ratios=[2, 2, 2, 2])

# Display prompt
print(f"Prompt: {data['Prompt']}")

# Display Noise
print(f"Noise: {data['Noise']}")

# Display K value (which_folder)
print(f"K Value (which_folder): {data['which_folder']}")

# Display Distance
print(f"Distance: {data['Distance']}")

# Load and display images
image_paths = [data['cache'], data['nocache'], data['retrieve']]
images = [Image.open(image_path) for image_path in image_paths]

# Titles for the images
image_titles = ['NIRVANA Image', 'Vanilla Image', 'GPTCache Image']

# Create a horizontal 1x3 grid for images with titles
for i, (image, title) in enumerate(zip(images, image_titles)):
    ax = plt.subplot(gs[i])
    ax.imshow(image, aspect='auto')
    ax.set_title(title, fontsize=12)  # Add image titles
    ax.axis('off')

# Create a tight layout
plt.tight_layout()

# Show the plot
plt.show()
```
