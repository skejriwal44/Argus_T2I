```python
import matplotlib.pyplot as plt
import numpy as np

# Data
tiny_time = [12.454516887664795, 11.537849426269531, 11.681047677993774, 11.5927255153656, 11.643336057662964]
small_time = [14.664707899093628, 14.70301103591919, 14.64878249168396, 15.566355228424072, 15.65803837776184]
sd_1_5_time = [21.65882897377014, 21.32762050628662, 21.143120527267456, 21.52659559249878, 22.83503222465515]
xl_time = [47.82641291618347, 47.0112783908844, 46.22635841369629, 46.64801502227783, 47.19372534751892]

tiny_size = 0.63
small_size = 2.32
sd_1_5_size = 3.44
xl_size = 5.14

categories = ['Tiny', 'Small', '1.5', 'XL']
time_means = [np.mean(tiny_time), np.mean(small_time), np.mean(sd_1_5_time), np.mean(xl_time)]
time_std = [3*np.std(tiny_time), 3*np.std(small_time), 3*np.std(sd_1_5_time), 3*np.std(xl_time)]

size_values = [tiny_size, small_size, sd_1_5_size, xl_size]

# Plotting
fig, ax1 = plt.subplots(figsize=(5, 5))

# Bar plot for time
bars = ax1.bar(categories, time_means, yerr=time_std, capsize=5, color='tab:blue', hatch = '/', alpha=0.7, edgecolor='black', label='Load Time \non GPU')
ax1.set_xlabel('SD Models', fontsize=24, color='black')
ax1.set_ylabel('Time (s)', color='blue', fontsize=22)
ax1.tick_params(axis='y', which='both', labelsize=24, colors='blue')
ax1.tick_params(axis='x', which='both', labelsize=22, colors='black')

ax1.yaxis.set_major_locator(plt.MaxNLocator(integer=True, prune='both', nbins=6))

# Create a second y-axis for the size
ax2 = ax1.twinx()
ax2.plot(categories, size_values, color='lightcoral', marker='o', markersize=8, label='Model card \nsize on disk')
ax2.set_ylabel('Size (in GB)', color='maroon', fontsize=26)
ax2.set_ylim(0,6)
ax2.tick_params(axis='y', which='both', labelsize=24, colors='maroon')

# ax2.yaxis.set_major_locator(plt.MaxNLocator(integer=True, prune='both', nbins=5))


# Add legend
# Add legends with proper positioning
ax1.legend(loc='upper left', fontsize=15)
ax2.legend(loc='upper left', bbox_to_anchor=(0, 0.8), fontsize=15)


# Title
# plt.title('Time and Size Comparison', fontsize=11)
plt.tight_layout()
# Show the plot
plt.savefig('load_time_size.pdf')
```
