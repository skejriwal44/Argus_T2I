## Low load vs High load

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Given data
load_types = ['Low Load', 'High Load']
throughput_accuscale_no_batch = [52.438164, 306.463038]
throughput_accuscale_batching = [47.614372, 368.242658]
latency_accuscale_no_batch = [13.4, 17.75]
latency_accuscale_batching = [6.95, 17.1]
slo_accuscale_no_batch = [0.035, 0.18]
slo_accuscale_batching = [0.04, 0.08]

# Set the figure size
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 3))

# Calculate the bar width
bar_width = 0.08
bar_positions = np.arange(len(load_types))/4

# Plotting the bars for throughput
axes[0].bar(bar_positions - bar_width/2, throughput_accuscale_no_batch, width=bar_width, hatch='/', color='tab:blue', edgecolor='black', label='AccuScale-with-Proteus-Batching')
axes[0].bar(bar_positions + bar_width/2, throughput_accuscale_batching, width=bar_width, hatch='\\', color='coral', edgecolor='black', label='AccuScale-with-Switch-Batching')

# Add labels, title, and legend
axes[0].set_ylabel('Avg Throughput', fontsize=18)
axes[0].set_xticks(bar_positions)
axes[0].set_xticklabels(load_types, fontsize=18)
axes[0].tick_params(axis='y', labelsize=14)


# Plotting the bars for latency
axes[1].bar(bar_positions - bar_width/2, latency_accuscale_no_batch, width=bar_width, hatch='/', color='tab:blue', edgecolor='black', label='AccuScale-with-Proteus-Batching')
axes[1].bar(bar_positions + bar_width/2, latency_accuscale_batching, width=bar_width, hatch='\\', color='coral', edgecolor='black', label='AccuScale-with-Switch-Batching')

# Add labels, title, and legend
axes[1].set_ylabel('Avg Latency', fontsize=18)
axes[1].set_xticks(bar_positions)
axes[1].set_xticklabels(load_types, fontsize=18)
axes[1].tick_params(axis='y', labelsize=14)
dash_len = 5
space = 4

# Add a second y-axis for SLO
axes2_0 = axes[0].twinx()
axes2_1 = axes[1].twinx()

# Plotting the lines for SLO
axes2_0.plot(bar_positions, slo_accuscale_no_batch, marker='o', linestyle='-', color='purple', markersize=10, label='SLO AccuScale-with-Proteus-Batching')
axes2_0.plot(bar_positions, slo_accuscale_batching, marker='^', linestyle='--',dashes=(dash_len, space), color='black', markersize=10, label='SLO AccuScale-with-Switch-Batching')

axes2_1.plot(bar_positions, slo_accuscale_no_batch, marker='o', linestyle='-', color='purple', markersize =10, label='SLO AccuScale-with-Proteus-Batching')
axes2_1.plot(bar_positions, slo_accuscale_batching, marker='^', linestyle='--',dashes=(dash_len, space), color='black', markersize =10, label='SLO AccuScale-with-Switch-Batching')

# Add labels and legend for the second y-axis
# axes2_0.set_ylabel('SLO Violations', fontsize=18, color='black', rotation=270)
# axes2_1.set_ylabel('SLO Violations', fontsize=18, color='black', rotation=270)

axes2_0.tick_params(axis='y', labelsize=14, labelcolor='black')
axes2_1.tick_params(axis='y', labelsize=14, labelcolor='black')

axes2_0.yaxis.get_major_ticks()[0].label1.set_visible(False)
axes2_1.yaxis.get_major_ticks()[0].label1.set_visible(False)

# Adjust y-axis limits for better visualization
axes2_0.set_ylim(-0.15, 0.2)
axes2_1.set_ylim(-0.15, 0.2)

axes[0].yaxis.set_major_locator(plt.MaxNLocator(8))
axes[1].yaxis.set_major_locator(plt.MaxNLocator(8))

axes2_1.yaxis.set_major_locator(plt.MaxNLocator(5))
axes2_0.yaxis.set_major_locator(plt.MaxNLocator(5))

# Remove grid lines
axes[0].grid(False)
axes[1].grid(False)
axes2_0.grid(False)
axes2_1.grid(False)


# Adjust layout
plt.tight_layout()

# Show the plot
plt.savefig('thr_latency_slo_load.png')
```
